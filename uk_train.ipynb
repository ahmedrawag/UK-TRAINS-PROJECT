**Which purchase method has the highest sales volume, online or from the station?**



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
ticket_counts = df['Purchase Type'].value_counts().reset_index()
ticket_counts.columns = ['Purchase Type', 'Total Tickets']
plt.figure(figsize=(8, 5))
sns.barplot(x='Purchase Type', y='Total Tickets', data=ticket_counts, hue='Purchase Type', palette='Blues', legend=False)
plt.xlabel("Purchase Type", fontsize=12)
plt.ylabel("Number of Tickets", fontsize=12)
plt.title("Ticket Sales by Purchase Type (Online vs. Station)", fontsize=14)
for i, v in enumerate(ticket_counts['Total Tickets']):
    plt.text(i, v + 50, str(v), ha='center', fontsize=12, fontweight='bold')
plt.show()
```

**Which are the most frequently used train stations?**



```python
ticket_counts = df['Departure Station'].value_counts().reset_index()
ticket_counts
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
station_counts = df['Departure Station'].value_counts().reset_index()
station_counts.columns = ['Departure Station', 'Total Tickets']
plt.figure(figsize=(10, 5))
sns.barplot(x='Departure Station', y='Total Tickets', data=station_counts, hue='Departure Station', palette='Blues', width=0.5, legend=False)
plt.xlabel("Departure Station", fontsize=12)
plt.ylabel("Number of Tickets", fontsize=12)
plt.title("Most Frequently Used Train Stations", fontsize=14)
plt.xticks(rotation=45, ha='right')
for i, v in enumerate(station_counts['Total Tickets']):
    plt.text(i, v + 50, str(v), ha='center', fontsize=10, fontweight='bold')
plt.show()
```

**What are the top 10 most traveled-to train stations?**



```python
import pandas as pd
import matplotlib.pyplot as plt
station_counts = df['Arrival Destination'].value_counts().reset_index()
station_counts.columns = ['Arrival Destination', 'Total Tickets']
top_10_stations = station_counts.nlargest(10, 'Total Tickets')
plt.figure(figsize=(12, 6))
plt.fill_between(range(len(top_10_stations)), top_10_stations['Total Tickets'], color='royalblue', alpha=0.5)
plt.plot(range(len(top_10_stations)), top_10_stations['Total Tickets'], color='blue', marker='o', linestyle='dashed')
plt.xticks(range(len(top_10_stations)), top_10_stations['Arrival Destination'], rotation=45, ha='right')
plt.xlabel("Arrival Destination", fontsize=12)
plt.ylabel("Number of Tickets", fontsize=12)
plt.title("Top 10 Most Traveled-To Train Stations ", fontsize=14)
plt.show()
```
